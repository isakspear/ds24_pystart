```python
axis: The scale lines on the plotting area(y and x). 
axes: the plotting are within the container.
figure:overall container
Artist: the visual elements drawn on the plot.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
#pyplot

import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(-4, 4, 400)  
y = x ** 2                    

# Create the plot
plt.figure(figsize=(8, 6))    
plt.plot(x, y, label='y = x^2', color='green', linewidth=2)

# title and labels
plt.title('Plot of y = x^2', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# legend
plt.legend()


plt.grid(True, linestyle='--', alpha=0.7)

# Display plot
plt.show()
```

```python
# Object-Oriented Approach:

import numpy as np
import matplotlib.pyplot as plt  


x = np.linspace(-4, 4, 400)  
y = x ** 2                    


fig, ax = plt.subplots(figsize=(8, 6))  


ax.plot(x, y, label='y = x^2', color='blue', linewidth=2)  


ax.set_title('Plot of y = x^2', fontsize=16)  
ax.set_xlabel('x', fontsize=14)             
ax.set_ylabel('y', fontsize=14)             


ax.legend()


ax.grid(True, linestyle='--', alpha=0.7)  


plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt


# Data for scatter plot
np.random.seed(15)                     
random_data_x = np.random.randn(1000)  
random_data_y = np.random.randn(1000)  


print(f"Scatter Plot Data - random_data_x: {random_data_x[:5]}")
print(f"Scatter Plot Data - random_data_y: {random_data_y[:5]}")

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())         # List of fruit names
values = list(fruit_data.values())      # Corresponding values


print(f"Bar Plot Data - names: {names}")
print(f"Bar Plot Data - values: {values}")


fig, axs = plt.subplots(1, 2, figsize=(14, 6))  



if random_data_x.size == 0 or random_data_y.size == 0:
    print("Error: Scatter plot data arrays are empty.")
else:
    
    scatter = axs[0].scatter(random_data_x, random_data_y, 
                             alpha=0.5,          # Transparency of markers
                             edgecolors='w',     # White edges around markers
                             linewidth=0.5,      # Width of marker edges
                             s=20,               # Size of markers
                             color='blue', 
                             zorder=2)           
    
   
    axs[0].set_title('Scatter Plot of Random Data', fontsize=16)
    axs[0].set_xlabel('Random X', fontsize=14)
    axs[0].set_ylabel('Random Y', fontsize=14)
    
    # Add grid
    axs[0].grid(True, linestyle='--', alpha=0.7, zorder=1)



if not names or not values:
    print("Error: Bar plot data is empty.")
else:
    
    bar = axs[1].bar(names, values, 
                     color=['purple', 'red', 'orange', 'yellow', 'green'],
                     edgecolor='black')
    
    
    axs[1].set_title('Fruit Count Bar Plot', fontsize=16)
    axs[1].set_xlabel('Fruit', fontsize=14)
    axs[1].set_ylabel('Count', fontsize=14)
    
    
    axs[1].grid(axis='y', linestyle='--', alpha=0.7)



plt.tight_layout()  
plt.show()        
```
